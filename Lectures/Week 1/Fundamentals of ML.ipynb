AI (Artificial Intelligence) is the broad discipline of creating intelligent machines<br>
ML (Machine Learning) refers to systems that can learn from experience<br>
DL (Deep Learning) refers to systems that learn from experience on large data sets<br>
NN (Neural Nets) are models of human neural networks that are designed to help computers learn<br>

DL is a data-driven modelling approach, which "learns the features", but what features?<br>
Complex models with millions/billions of parameters that need to be tuned

**Domain Shift** occurs when the source data distribution is different (but related) to the target data distribution, IE training your model on data and trying to use it on a similar dataset with some different feature(s) won't work due to domain shift

**Summary** <br>
DL is a subfield of ML, which is a subfield of AI<br>
DL models can learn the features from the data<br>
DL requires large datasets for training and scales better with more data
