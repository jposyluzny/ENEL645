```
def sum2(x,y,squared=False):
  if squared:
    s = (x + y)**2
  else:
    s = (x + y)
  return s
```

```
print('sum2(2,3):', sum2(2,3))
print('sum2(2,3,False):', sum2(2,3,False))
print('sum2(2,3,True):', sum2(2,3,True))
print('sum2(2,3,squared=True):', sum2(2,3,squared=True))
```
